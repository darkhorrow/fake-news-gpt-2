# Generación del dataset final a usar

```
import os
import json
from pathlib import  Path
```

```
i = 0

with open('text_file.txt', 'w') as text_file:
  for root, dirs, files in os.walk("LAS PALMAS"):
    if i < 50:
      for name in files:
          if name.endswith((".json")):
              json_file = json.load(open(os.path.join(root, name), 'r'))
              text_file.write(json_file['body']['raw_text'])
              text_file.write('\n')
              i = i + 1
    else:
      break
```

```
from google_trans_new import google_translator  
translator = google_translator()  

with open('text_file.txt', 'r') as file:
  with open('text_file_translation.txt', 'w') as save_file:
    for line in file.readlines():
      translate_line = translator.translate(line, lang_tgt='en')
      save_file.write(translate_line)
```
